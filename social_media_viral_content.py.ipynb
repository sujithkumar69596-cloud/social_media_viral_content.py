```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df.fillna(0, inplace=True)
```

```python
df['total_engagement'] = df['likes'] + df['comments'] + df['shares']
```

```python
df.loc[df['total_engagement'].idxmax()]
```

```python
df.groupby('content_type')['total_engagement'].mean()
```

```python
df.groupby('content_type')['total_engagement'].sum().idxmax()
```

```python
np.mean(df['total_engagement'])
```

```python
np.max(df['total_engagement'])
```

```python
np.min(df['total_engagement'])
```

```python
df.groupby('content_type')['total_engagement'].mean().plot(kind='bar')
plt.xlabel("Content Type")
plt.ylabel("Average Engagement")
plt.title("Average Engagement by Content Type")
plt.show()
```

```python
df['post_datetime'] = pd.to_datetime(df['post_datetime'])
```

```python
df.groupby('post_datetime')['total_engagement'].sum().plot()
plt.xlabel("Date")
plt.ylabel("Total Engagement")
plt.title("Engagement Trend Over Time")
plt.show()
```
